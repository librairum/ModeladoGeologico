```python
import gempy as gp
import gempy_viewer as gpv
from gempy_engine.core.data.stack_relation_type import StackRelationType
```

```python
# Crear un modelo geológico con capas deformadas (plegadas/inclinadas)
data = gp.create_geomodel(
    project_name='deformed_layers',
    extent=[0, 2000, 0, 1000, 0, 1200],
    refinement=6,
    importer_helper=gp.data.ImporterHelper(
        path_to_orientations="input_data/extended_model2_orientations.csv",
        path_to_surface_points="input_data/extended_model2_surface_points.csv"
    )
)

# Mapear series geológicas a superficies
gp.map_stack_to_surfaces(
    gempy_model=data,
    mapping_object={
        "Deformed_Series": ('rock1', 'rock2', 'rock3', 'rock4', 'rock5')
    }
)

# Computar el modelo geológico
gp.compute_model(data)
geo_data = data

# Visualizar en 2D para analizar la deformación de las capas
gpv.plot_2d(geo_data, direction='y', show_data=True)
```
